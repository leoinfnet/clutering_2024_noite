```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import seaborn as sns
```

```python
X, y = make_moons(n_samples=1000, noise=0.06, random_state=42)
```

```python
fig, axs= plt.subplots(2,2, figsize=(12,10))
methods = ['single','complete','average', 'ward']
titles = ['Single Linkage', 'Complete Linkage', 'Average Linkage' ,'Ward']

for ax, method, title in zip(axs.flatten(), methods, titles):
    clustering = AgglomerativeClustering(n_clusters=2, linkage=method)
    y_pred = clustering.fit_predict(X)
    sns.scatterplot(x=X[:,0],y=X[:,1], hue=y_pred, ax=ax , palette='viridis')
    ax.set_title(title)
    ax.set_xlabel(" Feature 1 ")
    ax.set_ylabel( " Feature 2 " )
plt.tight_layout()
plt.show()
```

```python
np.random.seed(42)
X = np.random.rand(100,2)
```

```python
fig, axs= plt.subplots(2,2, figsize=(12,10))
methods = ['single','complete','average', 'ward']
titles = ['Single Linkage', 'Complete Linkage', 'Average Linkage' ,'Ward']

for ax, method, title in zip(axs.flatten(), methods, titles):
    clustering = AgglomerativeClustering(n_clusters=5, linkage=method)
    y_pred = clustering.fit_predict(X)
    sns.scatterplot(x=X[:,0],y=X[:,1], hue=y_pred, ax=ax , palette='plasma')
    ax.set_title(title)
    ax.set_xlabel(" Feature 1 ")
    ax.set_ylabel( " Feature 2 " )
plt.tight_layout()
plt.show()
```

```python
Z = linkage(X,method='ward')

plt.figure(figsize= (12,8))
plt.subplot(1,2,1)
dendrogram(Z)

plt.title('Dendrogram')
plt.xlabel("Amostras")
plt.ylabel("distancias")

clustering = AgglomerativeClustering(n_clusters=2, linkage='ward')
y_pred = clustering.fit_predict(X)

plt.subplot(1,2,2)
sns.scatterplot(x=X[:,0], y=X[:,1], hue=y_pred, palette='plasma', legend=False)
plt.title("Clusterização Aglomerativa(Ward)")
plt.tight_layout()
plt.show()




```

```python
# Gerar dados em forma de lua
X, _ = make_moons(n_samples=200, noise=0.1, random_state=42)

# Calcular a hierarquia usando o método Ward
Z = linkage(X, method='ward')

# Configuração da figura
plt.figure(figsize=(12, 8))

# Plotar o dendrograma
plt.subplot(1, 2, 1)
dendrogram(Z)
plt.title('Dendrograma - Método Ward')
plt.xlabel('Amostras')
plt.ylabel('Distância')

# Aplicar clusterização aglomerativa
clustering = AgglomerativeClustering(n_clusters=2, linkage='ward')
y_pred = clustering.fit_predict(X)

# Plotar resultados dos clusters
plt.subplot(1, 2, 2)
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y_pred, palette='plasma', legend=False)
plt.title('Clusterização Aglomerativa (Ward) com Dados em Lua')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()
```

```python
# Gerar dados em forma de lua
X, _ = make_moons(n_samples=200, noise=0.1, random_state=42)

# Calcular a hierarquia usando o método Ward
Z = linkage(X, method='single')

# Configuração da figura
plt.figure(figsize=(12, 8))

# Plotar o dendrograma
plt.subplot(1, 2, 1)
dendrogram(Z)
plt.title('Dendrograma - Método Single')
plt.xlabel('Amostras')
plt.ylabel('Distância')

# Aplicar clusterização aglomerativa
clustering = AgglomerativeClustering(n_clusters=2, linkage='ward')
y_pred = clustering.fit_predict(X)

# Plotar resultados dos clusters
plt.subplot(1, 2, 2)
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y_pred, palette='plasma', legend=False)
plt.title('Clusterização Aglomerativa (Single) com Dados em Lua')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()
```
